```python
import sys
import time

def to_binary(n):
    """Convert number to binary using basic algorithm."""
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary

def to_hexadecimal(n):
    """Convert number to hexadecimal using basic algorithm."""
    hex_chars = "0123456789ABCDEF"
    if n == 0:
        return "0"
    hexadecimal = ""
    while n > 0:
        hexadecimal = hex_chars[n % 16] + hexadecimal
        n //= 16
    return hexadecimal

def process_file(file_path):
    """Read numbers from file, convert them, and handle errors."""
    results = []
    start_time = time.time()

    try:
        with open(file_path, "r") as file:
            lines = file.readlines()

        for line in lines:
            line = line.strip()
            if line.isdigit():
                num = int(line)
                binary = to_binary(num)
                hexadecimal = to_hexadecimal(num)
                results.append(f"{num}: Binary={binary}, Hex={hexadecimal}")
            else:
                results.append(f"Invalid data: {line}")

    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return

    end_time = time.time()
    execution_time = end_time - start_time
    results.append(f"Execution Time: {execution_time:.6f} seconds")


    for res in results:
        print(res)


    output_file = "/content/drive/MyDrive/P2/ConvertionResults.txt"
    with open(output_file, "w") as file:
        file.write("\n".join(results))

    print(f"Results saved to {output_file}")


file_path = "/content/drive/MyDrive/P2/TC1.txt"
process_file(file_path)
```

```python
import time
import os

def to_binary(n):
    """Convert number to binary using basic algorithm."""
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary

def to_hexadecimal(n):
    """Convert number to hexadecimal using basic algorithm."""
    hex_chars = "0123456789ABCDEF"
    if n == 0:
        return "0"
    hexadecimal = ""
    while n > 0:
        hexadecimal = hex_chars[n % 16] + hexadecimal
        n //= 16
    return hexadecimal

def process_files(directory, filenames):
    """Process multiple files and convert numbers."""
    output_file = os.path.join(directory, "ConvertionResults.txt")
    results = []
    start_time = time.time()

    for filename in filenames:
        file_path = os.path.join(directory, filename)
        results.append(f"Processing file: {filename}\n")

        try:
            with open(file_path, "r") as file:
                lines = file.readlines()

            for line in lines:
                line = line.strip()
                if line.isdigit():
                    num = int(line)
                    binary = to_binary(num)
                    hexadecimal = to_hexadecimal(num)
                    results.append(f"{num}: Binary={binary}, Hex={hexadecimal}")
                else:
                    results.append(f"Invalid data: {line}")

        except FileNotFoundError:
            results.append(f"Error: File '{filename}' not found.")
            continue

    end_time = time.time()
    execution_time = end_time - start_time
    results.append(f"Execution Time: {execution_time:.6f} seconds\n")


    for res in results:
        print(res)


    with open(output_file, "w") as file:
        file.write("\n".join(results))

    print(f"Results saved to {output_file}")


directory = "/content/drive/MyDrive/P2"
filenames = ["TC1.txt", "TC2.txt", "TC3.txt", "TC4.txt"]


process_files(directory, filenames)
```
