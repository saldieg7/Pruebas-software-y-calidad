```python
import time
import os
from google.colab import drive
```

```python
# Lista de archivos de texto
file_names = ["TC1.txt", "TC2.txt", "TC3.txt", "TC4.txt", "TC5.txt", "TC6.txt", "TC7.txt"]
```

```python
 #dirección de carpeta
folder_path = "/content/drive/MyDrive/P1/"
```

```python
# Se leen los numeros del archivo txt
def read_numbers_from_file(file_path):
    """Reads numbers from a file, ignoring invalid entries."""
    numbers = []
    errors = 0
    try:
        with open(file_path, 'r') as file:
            for line in file:
                try:
                    numbers.append(float(line.strip()))
                except ValueError:
                    errors += 1
    except Exception as e:
        print(f"Error opening file {file_path}: {e}")
    return numbers, errors
```

```python
# Funciones que calculan las estadisticas
def compute_mean(numbers):
    """Calculates mean without built-in functions."""
    total = sum(numbers)
    return total / len(numbers) if numbers else 0

def compute_median(numbers):
    """Calculates median manually."""
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n == 0:
        return 0
    mid = n // 2
    if n % 2 == 0:
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:
        return sorted_numbers[mid]

def compute_mode(numbers):
    """Calculates mode manually."""
    frequency = {}
    for num in numbers:
        frequency[num] = frequency.get(num, 0) + 1
    max_count = max(frequency.values(), default=0)
    modes = [num for num, count in frequency.items() if count == max_count]
    return modes[0] if len(modes) == 1 else modes

def compute_variance(numbers, mean):
    """Calculates variance manually."""
    total = sum((num - mean) ** 2 for num in numbers)
    return total / len(numbers) if numbers else 0

def compute_standard_deviation(variance):
    """Calculates standard deviation manually."""
    return variance ** 0.5

# Se procesa cada archivo de l alista
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)


    print(f"Checking file: {file_path}")


    if not os.path.exists(file_path):
        print(f"Error: File {file_name} not found. Skipping...\n")
        continue

    print(f"Processing {file_name}...")


    start_time = time.time()


    numbers, invalid_count = read_numbers_from_file(file_path)

    # Resultados de interes
    mean = compute_mean(numbers)
    median = compute_median(numbers)
    mode = compute_mode(numbers)
    variance = compute_variance(numbers, mean)
    std_dev = compute_standard_deviation(variance)


    execution_time = time.time() - start_time


    output = f"""
    Descriptive Statistics for {file_name}:
    ----------------------------------------
    Mean: {mean}
    Median: {median}
    Mode: {mode}
    Variance: {variance}
    Standard Deviation: {std_dev}
    Invalid Entries Ignored: {invalid_count}
    Execution Time: {execution_time:.6f} seconds
    """


    print(output)


    results_file = "/content/drive/MyDrive/P1/StatisticsResults.txt"
    with open(results_file, 'a') as file:
        file.write(output)
        file.write("\n\n")

print("All files processed. Results saved in 'StatisticsResults.txt'.")
```
