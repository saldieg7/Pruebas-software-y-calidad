```python
import time
import os

def count_words_in_file(file_path):
    try:

        with open(file_path, 'r') as file:
            text = file.read()

        words = ''.join([char if char.isalnum() else ' ' for char in text]).split()

        word_count = {}

        for word in words:
            word = word.lower()
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1

        return word_count

    except Exception as e:
        print(f"Error while reading file {file_path}: {e}")
        return None

def save_results(word_count, output_file_path, elapsed_time):
    try:

        with open(output_file_path, 'w') as output_file:
            output_file.write("Word Count Results:\n")
            for word, count in sorted(word_count.items()):
                output_file.write(f"{word}: {count}\n")
            output_file.write(f"\nElapsed Time: {elapsed_time} seconds\n")
    except Exception as e:
        print(f"Error while writing results to {output_file_path}: {e}")

def main():
    start_time = time.time()

    # (Google Colab path)
    input_file_path = '/content/drive/MyDrive/P3/TC1.txt'
    output_file_path = '/content/drive/MyDrive/P3/WordCountResults.txt'

    # Númeor de palabras en el archivo de txt
    word_count = count_words_in_file(input_file_path)

    if word_count is not None:

        for word, count in sorted(word_count.items()):
            print(f"{word}: {count}")

        # Save the results to a file
        elapsed_time = time.time() - start_time
        save_results(word_count, output_file_path, elapsed_time)


        print(f"\nElapsed Time: {elapsed_time} seconds")
    else:
        print("An error occurred during word counting.")

if __name__ == '__main__':
    main()
```
